# USE THIS TO SEE OUTPUTS OF THE OTHER PYTHON FILES, IF YOU CANNOT EXECUTE THEM FOR EXAMPLE

```python
import sys
import io

# Function to capture output of a script execution
def run_script(script_name):
    print(f"\n### Running {script_name} ###\n")
    old_stdout = sys.stdout  # Save the original stdout
    sys.stdout = io.StringIO()  # Redirect stdout to capture prints

    try:
        %run {script_name}
    except Exception as e:
        print(f"Error executing {script_name}: {e}")

    output = sys.stdout.getvalue()  # Get captured output
    sys.stdout = old_stdout  # Restore original stdout
    print(output)  # Display the output

# Run scripts and display outputs
run_script("baseline_model.py")
run_script("comp_models.py")
run_script("hyperparam_tuning.py")
```
